ViT(vision transformer) and CLIP(contrastive language-image pretraining) are two popular transformer-based models used for image processing tasks.

```python
from transformers import CLIPModel, CLIPProcessor
```

```python
from PIL import Image
```

```python
path = r"/home/abhishek/GitHub_Projects/Full-Stack-GenAI-Bootcamp-1.0/Class-08-18-Apr-SOTA-Transformer-Embedding-Methods/image/flower.jpg"
```

```python
image = Image.open(path)
```

```python
image.show()
```

```python
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

```python
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
```

```python
inputs =processor(images=image, return_tensors="pt")
```

```python
import torch
```

```python
# Get embedding
with torch.no_grad():
    image_embedding = model.get_image_features(**inputs)
```
